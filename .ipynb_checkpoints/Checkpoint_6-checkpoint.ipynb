```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import seaborn as sns
import category_encoders as ce
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
```

```python
# Descargar el conjunto de datos
url = "https://data.irazusystems.site/fruit_classification_dataset.csv"
# Nombres descriptivos para las columnas

# Cargar los datos
df = pd.read_csv(url)

# Vista inicial del conjunto de datos
print("Primeras filas del conjunto de datos:")
df.head()
```

```python
# Verificar valores faltantes
print("\nValores faltantes por columna:")
df.isnull().sum()
```

```python
# Codificar variables categóricas
for column in ["shape", "color", "taste", "fruit_name"]:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
```

```python
# class es la variable objetivo
X = df.drop(['fruit_name'], axis=1)
y = df['fruit_name']
```

## Árbol de decisiones

```python
# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo
tree_model = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_model.fit(X_train, y_train)

# Predicción
y_pred = tree_model.predict(X_test)

# Evaluar
print("Precisión:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Bosque Aleatorio

```python
# Modelo de bosque aleatorio
rfc = RandomForestClassifier(n_estimators=10,random_state=0)

# ajustar y predecir
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)

# Evaluar precisión
print('Precisión con 10 árboles: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
